## Предобработка данных

### Обработка пропусков

#### Категориальные переменные

Заполним пропуски в столбце *is_apartment*.

```python
df['is_apartment'] = df['is_apartment'].fillna(False)
df['is_apartment'].unique()
```

Изучим подробнее пропуски в *locality_name*.

```python
df[df['locality_name'].isna()]
```

Мы получили таблицу из 49 строк. Очевидно, что по характеру пропуски в *locality_name* полностью случайные, т.к. для большинства объектов есть данные об их географическом положении, о чём свидетельствуют заполненные столбцы *airports_nearest, cityCenters_nearest, parks_around3000, parks_nearest, ponds_around3000, ponds_nearest*.
Попробуем восстановить эти пропуски, используя имеющуюся информацию в *airports_nearest* и *cityCenters_nearest*, как двух наиболее заполненных столбцах.

Напишем функцию, которая будет искать в исходной таблице объекты, расположенные максимально близко к целевому объекту, но не дальше чем 3 км от него. Шагом увеличения радиуса возьмём 10 метров.

```python
def find_locality(row):
    
    locality = row['locality_name']
    if pd.isna(locality):
        airports_nearest = row['airports_nearest']
        cityCenters_nearest = row['cityCenters_nearest']
        
        radius = 0
        while pd.isna(locality) and radius<3000:
            try:
                locality = df[(df['airports_nearest']>=airports_nearest-radius)&
                              (df['airports_nearest']<=airports_nearest+radius)&
                              (df['cityCenters_nearest']>=cityCenters_nearest-radius)&
                              (df['cityCenters_nearest']<=cityCenters_nearest+radius)&
                              ~(df['locality_name'].isna())]['locality_name'].iloc[0]
            except:
                radius+=10
 
    return locality
```

```python
df['locality_name'] = df.apply(find_locality, axis=1)
df[df['locality_name'].isna()]
```

После выполнения функции всё равно остались пропуски в *locality_name* - это объекты, по которым вообще нет никаких географических данных. Обозначим, что у них город неизвестен.

```python
df['locality_name'] = df['locality_name'].fillna('Неизвестно')
df['locality_name'].isna().sum()
```

Пропуски в категориальных переменных *is_apartment* и *locality_name* заполнены.

#### Количественные переменные

Проверим нули в *rooms*.

```python
df[df['rooms']==0]['studio'].value_counts()
```

```python
df[(df['rooms']==0)&(~df['studio'])]['open_plan'].value_counts()
```

Видим, что комнаты действительно отсутствуют только в случае студии или свободной планировки, значит, это не ошибка.

Заменим пропуски на 0 в *kitchen_area*. Для этого напишем функцию, которая будет обрабатывать столбец *kitchen_area* следующим образом:
1. Если в *kitchen_area* пропуск, а в *studio* и *open_plan* - *True* - заменяем площадь кухни на 0.
2. Если же это не студия и/или свободная планировка, площадь кухни рассчитываем как медианную долю площади кухни у аналогичных по комнатности и площади квартир от общей площади. Площадь разобъём на категории по 10 кв.метров.
3. Если у квартиры нет аналогичных ей по площади и комнатности квартир, площадь кухни рассчитываем как медианную долю площади кухни у квартир с той же комнатностью от общей площади.

```python
def fill_kitchen(row):
    kitchen_area = row['kitchen_area'] 
    if pd.isna(kitchen_area):
        if row['studio'] or row['open_plan']:
            kitchen_area = 0
        else:
            df2 = df[~(df['kitchen_area'].isna())&
                     (df['rooms']==row['rooms'] )&
                     (round(df['total_area'],-1)==round(row['total_area'],-1))]
            if len(df2):
                kitchen_area = (df2['kitchen_area']/df2['total_area']).median()*row['total_area']
            else:
                df3 = df[(df['rooms']==row['rooms'] )]
                kitchen_area = (df3['kitchen_area']/df3['total_area']).median()*row['total_area']
    return kitchen_area
```

```python
df['kitchen_area'] = df.apply(fill_kitchen, axis=1)
df['kitchen_area'].isna().sum()
```

В *kitchen_area* больше нет пропусков.

Избавимся от пропусков в *balcony*, а пропуски в *living_area* посчитаем как разность общей площади от площади кухни.

```python
df['balcony'] = df['balcony'].fillna(0)
df['living_area'] = df['living_area'].fillna(df['total_area']-df['kitchen_area'])
```

Проверим, что в *living_area* не появилось отрицательных значений.

```python
df['living_area'].describe()
```

Напишем функцию для заполнения пропусков в *ceiling_height*, которая будет работать следующим образом:
1. Если в *ceiling_height* пропуск, высоту потолка рассчитываем как медианную высоту у аналогичных по этажности домов того же района.
2. Если в районе нет домов с той же этажностью, берем медиану по всем домам той же этажности.

```python
def fill_ceiling_height(row):
    ceiling_height = row['ceiling_height']
    if pd.isna(ceiling_height):
        df2 = df[~(df['ceiling_height'].isna())&
                 (df['floors_total']==row['floors_total'] )&
                 (df['locality_name']==row['locality_name'])]
        if len(df2):
            ceiling_height = df2['ceiling_height'].median()
        else:
            df3 = df[(df['floors_total']==row['floors_total'] )]
            ceiling_height = df3['ceiling_height'].median()
    return ceiling_height
        
```

```python
df['ceiling_height'] = df.apply(fill_ceiling_height, axis=1)
df[df['ceiling_height'].isna()]
```

Остались незаполненными 78 пропусков высоты потолка. Это квартиры в домах, по которым неизвестна этажность. Заменим эти пропуски на медианное значение высоты потолка по всему датафрейму.

```python
df['ceiling_height'] = df['ceiling_height'].fillna(df['ceiling_height'].median())
df['ceiling_height'].isna().sum()
```

Пропусков в ceiling_height больше нет. 

Заменить пропуски в floors_total средним значением не представляется возможным, поэтому оставим их пустыми.

Значения в столбцах *floors_total* и *balcony* были с плавающей точкой, мы заменили их на целочисленные для удобства обработки и экономии памяти. Даты в *first_day_exposition* переведены к типу данных *datetime*.

## Расчёты и добавление результатов в таблицу

Добавим в таблицу некоторые дополнительно рассчитанные данные. Начнем с цены квадратного метра.

```python
df['meter_price'] = df['last_price']/df['total_area']
```

Отдельно вынесем день недели, месяц и год публикации объявления.

```python
df['weekday'] = df['first_day_exposition'].dt.weekday
df['month'] = df['first_day_exposition'].dt.month
df['year'] = df['first_day_exposition'].dt.year
```

Категоризируем значения этажа. Для этого напишем функцию, которая будет однозначно определять - этаж первый, последний или другой.

```python
def floor(row):
    floor = row['floor']
    floors_total = row['floors_total']
    if floor == 1:
        return 'Первый'
    if floor == floors_total:
        return 'Последний'
    return 'Другой'

df['floor_type'] = df.apply(floor, axis=1)
```

Добавим в таблицу соотношение жилой и общей площади, а также отношение площади кухни к общей.

```python
df['living_area_ratio'] = df['living_area']/df['total_area']
df['kitchen_area_ratio'] = df['kitchen_area']/df['total_area']
```

### Изучение зависимости показателей друг от друга

Изучим зависимость цены от площади, числа комнат, удалённости от центра, этажа, даты размещения объявления и узнаем, какие факторы больше всего влияют на стоимость квартиры.

Посчитаем коэффициент корреляции Пирсона и отобразим зависимости на графиках.

```python
print('Коэффициент корреляции Пирсона: {:.2}'.format(df['last_price'].corr(df['total_area'])))
df.plot(x='last_price',
        y='total_area',
        kind='scatter',
        title='Зависимость цены от площади',
        figsize=(10, 5),
        grid=True,
        alpha=0.1)
plt.show()
```

Согласно коэффициенту Пирсона и диаграмме разброса, цена достаточно сильно положительно линейно зависит от площади квартиры, что логично. 

```python
print('Коэффициент корреляции Пирсона: {:.2}'.format(df['last_price'].corr(df['rooms'])))
df.groupby('rooms')['last_price'].median().plot(x='rooms',
        y='last_price',
        kind='bar',
        title='Зависимость цены от числа комнат',
        figsize=(10, 5),
        grid=True)
plt.show()
```

Чем больше в квартире комнат, тем она дороже. Однако данный показатель не такой решающий, как площадь квартиры.

```python
print('Коэффициент корреляции Пирсона: {:.2}'.format(df['last_price'].corr(df['cityCenters_nearest'])))
df.plot(x='last_price',
        y='cityCenters_nearest',
        kind='scatter',
        title='Зависимость цены от удаленности от центра',
        figsize=(10, 5),
        grid=True,
        alpha=0.1)
plt.show()
```

Цена квартиры отрицательно зависит от её удалённости от центра, однако и на расстоянии 10-20 км от центра есть квартиры дешевле 7.5 миллионов рублей, поэтому данная зависимость выражена не так сильно, как предыдущие.

```python
df.groupby('floor_type')['meter_price'].median().plot(x='floor_type',
        y='meter_price',
        kind='bar',
        title='Зависимость цены метра от этажа',
        figsize=(5, 5),
        grid=True)
plt.show()
```

Квартиры на первом этаже значительно дешевле остальных. Самые дорогие квартиры в категории "Другие" - среди тех, что не на последнем и не на первом.

```python
print('Коэффициент корреляции Пирсона: {:.2}'.format(df['last_price'].corr(df['weekday'])))
df.groupby('weekday')['last_price'].median().plot(x='weekday',
        y='last_price',
        kind='bar',
        title='Зависимость цены от дня недели',
        figsize=(10, 5),
        grid=True)
plt.show()
```

```python
print('Коэффициент корреляции Пирсона: {:.2}'.format(df['last_price'].corr(df['month'])))
df.groupby('month')['last_price'].median().plot(x='month',
        y='last_price',
        kind='bar',
        title='Зависимость цены от месяца',
        figsize=(10, 5),
        grid=True)
plt.show()
```

Зависимости цены от дня недели и месяца публикации объявления не прослеживается.

```python
print('Коэффициент корреляции Пирсона: {:.2}'.format(df['last_price'].corr(df['year'])))
df.groupby('year')['last_price'].median().plot(x='year',
        y='last_price',
        kind='bar',
        title='Зависимость цены от года',
        figsize=(10, 5),
        grid=True)
plt.show()
```

```python
df.groupby('year')['last_price'].count().plot(x='year',
        y='last_price',
        kind='pie',
        title='Количество объявлений по годам',
        figsize=(10, 5),
        grid=True)
plt.show()
```

На первом графике 2014 год сильно выделяется на фоне остальных, однако нужно учитывать, что за этот год совсем мало данных, поэтому нельзя сделать вывод, что в 2014 году квартиры были сильно дороже. В целом же с 2016 года цены на недвижимость имеют тенденцию расти.

### Вывод

1. Сильнее всего цена зависит от площади - чем больше квартира занимает квадратных метров, тем она дороже. 
2. Комнтаность также влияет на цену, но более косвенно, т.к. не всегда большое количество комнат означает большое количество квадратных метров.

### Исследование центра Санкт-Петербурга

Исследуем предложение квартир в Санкт-Петербурге и выясним, как меняются цены в зависимости от удаленности от центра и в каких рамках находится центр.

Выделим в отельный датафрейм данные по Санкт-Петербургу, для которых известно расстояние до центра.

```python
df_spb = df[(df['locality_name']=="Санкт-Петербург")&(~df['cityCenters_nearest'].isna())]
df_spb.info()
```

График сильно меняется сразу после 8 километров, значит 0-8 километров и есть центр города.

Выделим сегмент квартир в центре и проанализируем эту территорию.

```python
df_center = df_spb[df_spb['km_to_center']<=8]
df_center.info()
```

В отличие от области, в центре популярнее двух- и трехкомнатные квартиры.

1. В среднем квартира в Санкт-Петребурге и Ленинградской области стоит 5 миллионов рублей, занимает 50 квадратных метров, имеет 1-2 комнаты и потолки высотой 2.7 метра. Средний срок продажи квартиры - 2-3 месяца. Квартиры на первом и последнем этажах дешевле.
2. Самая высокая цена квадратного метра среди крупных населенных пунктов - в Санкт-Петербурге и Пушкине, самая низкая - в Выборге.
3. В центре Санкт-Петербурга квартиры дороже и с более высокими потолками, а выбор просторных квартир (более 60 кв.метров) представлен шире. В центре чаще покупают 2- и 3-комнатные квартиры.
3. Сильнее всего цена квартиры зависит от площади. Вторым по значимости фактором является комнатность. Чем ближе квартира к центру, тем она дороже, но это не распространяется на квартиры в центральном районе.
